````python
import requests 
import bs4 as bs
import pandas as pd
import numpy as np
import urllib.request

source = requests.get("http://www.debates.org/index.php?page=debate-transcripts") 
soup = bs.BeautifulSoup(source.content, features='html.parser') 
links = []
title = []
for url in soup.find_all('a'):
    link = url.get('href')
    string = url.text
    if "The First " in string:
        print(string)
        print(link)
        links.append(link)
        title.append(string)
````

````python
from collections import Counter
import re
import pandas as pd

index = ["Debate char length", "war_count", "most_common_w", "most_common_w_count"]
df = pd.DataFrame(index=index, columns=title)
df = df.fillna(0)
dumb = 0

for link in links:
    source1 = requests.get(link) 
    soup1 = bs.BeautifulSoup(source1.content, features='html.parser')
    content = soup1.find("div", id="content-sm")
    ps = content.find('p').text
    df.iloc[0,dumb] = len(ps)
    counts = Counter(re.findall("war[\s.,!s]", ps.lower()))
    df.iloc[1,dumb] = sum(counts.values())
    psclean = re.sub("[^a-zA-Z]", " ",ps)
    psclean2 = psclean.lower().split()
    df.iloc[2,dumb] = Counter(psclean2).most_common(1)[0][0]
    df.iloc[3,dumb] = Counter(psclean2).most_common(1)[0][1]
    dumb = dumb + 1
````

````python
rawp2 = requests.get('http://people.sc.fsu.edu/~jburkardt/datasets/regression/')
soupp2 = bs.BeautifulSoup(rawp2.content,features='html.parser')
links2 = []
for url in soupp2.find_all('a')[6:33]:
    link2 = "http://people.sc.fsu.edu/~jburkardt/datasets/regression/" + url.get('href')
    links2.append(link2)

````

````python
authorlist = []

for link in links2:
    source = requests.get(link) 
    soup = bs.BeautifulSoup(source.content, features='html.parser')
    li = soup.prettify().split('\n')
    author = str(li[4])
    cleanauthor = re.sub("[^a-zA-Z]", "", author)
    authorlist.append(cleanauthor)
    
````

````python
dict1 = Counter(authorlist).most_common()
index1 = ["0", "1", "2", "3","4","5","6"]
title1 = ["Authors", "Counts"]
df = pd.DataFrame(index=index1, columns=title1)
df = df.fillna(0)
for i in range(0,7):
    df.iloc[i,0] = dict1[i][0]
    df.iloc[i,1] = dict1[i][1]
    
df
````
